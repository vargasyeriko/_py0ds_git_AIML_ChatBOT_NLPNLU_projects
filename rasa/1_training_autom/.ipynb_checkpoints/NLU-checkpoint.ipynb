```python
import pandas as pd

def append_nlu_file(dataframe, nlu_file_path):
    with open(nlu_file_path, 'a') as nlu_file:  # 'a' mode appends to the file
        current_intent = None
        for index, row in dataframe.iterrows():
            intent = row['Intent']
            example = row['Examples']

            if intent != current_intent:
                nlu_file.write('\n')
                nlu_file.write(f"- intent: {intent}\n")
                nlu_file.write(f"  examples: |\n")
                current_intent = intent

            nlu_file.write(f"    - {example}\n")

# Specify the path of your NLU file
nlu_file_path = 'nlu.yml'

# Append new entries to the NLU file from the DataFrame
append_nlu_file(df_int, nlu_file_path)

# Read and print the appended NLU content
with open(nlu_file_path, 'r') as nlu_file:
    nlu_content = nlu_file.read()

print(f"Appended to NLU file '{nlu_file_path}'.")
print("Updated NLU content:")
print(nlu_content)
```

```python
import pandas as pd
import yaml

def append_intents_to_yaml(df, file_name="output.yml"):
    # Define a custom representer for strings to always quote them
    def represent_str(dumper, data):
        if '\n' in data:
            return dumper.represent_scalar('tag:yaml.org,2002:str', data, style='|')
        return dumper.represent_scalar('tag:yaml.org,2002:str', data)
    
    yaml.add_representer(str, represent_str)
    
    # Ensure the DataFrame has the correct columns
    if 'Intent' not in df or 'Response' not in df:
        raise ValueError("DataFrame must contain 'Intent' and 'Response' columns")
    
    # Load existing YAML data
    try:
        with open(file_name, "r") as yaml_file:
            existing_data = yaml.safe_load(yaml_file) or {}
    except FileNotFoundError:
        existing_data = {}
    
    # Prepare the responses dictionary using a comprehension
    # Prefix intents with 'utter_' for responses and append to existing responses
    new_responses = {f"utter_{intent}": [{"text": text}] for intent, text in zip(df["Intent"], df["Response"])}
    existing_responses = existing_data.get("responses", {})
    existing_responses.update(new_responses)

    # Append new intents to existing intents
    existing_intents = existing_data.get("intents", [])
    new_intents = [intent for intent in df["Intent"] if intent not in existing_intents]
    existing_intents.extend(new_intents)

    # Update the existing YAML data structure
    existing_data.update({
        "version": "3.1",
        "intents": existing_intents,
        "responses": existing_responses,
        "session_config": {
            "session_expiration_time": 60,
            "carry_over_slots_to_new_session": True
        }
    })

    # Write the updated YAML data back to the file with text fields quoted
    with open(file_name, "w") as yaml_file:
        yaml.dump(existing_data, yaml_file, allow_unicode=True, default_flow_style=False, sort_keys=False, indent=2)

    return file_name

# Example DataFrame with intents and responses
df_example = pd.DataFrame({
    'Intent': ['new_intent_1', 'new_intent_2'],
    'Response': ['Response to new intent 1', 'Response to new intent 2']
})

# Call the function to append new intents and responses to the existing YAML file
file_path = append_intents_to_yaml(df_int, 'nlu.yml')  # Ensure to use the actual existing file name
```

```python
def append_df_to_nlu(df, nlu_file_path):
    # Read the existing NLU data
    with open(nlu_file_path, 'r', encoding='utf8') as file:
        nlu_data = yaml.safe_load(file) or {'nlu': []}

    # Check if 'nlu' is a key in the dictionary, as expected
    if 'nlu' not in nlu_data or not isinstance(nlu_data['nlu'], list):
        raise ValueError("The NLU file is not formatted as expected. 'nlu' key not found or not a list.")

    # Group by 'Intent' and combine 'Examples'
    grouped_df = df.groupby('Intent')['Examples'].apply(lambda x: "\n ".join(x)).reset_index()

    # Build new data structure
    new_data_structure = [{'intent': row['Intent'], 'examples': f"|\n {row['Examples']}"} for index, row in grouped_df.iterrows()]

    # Append new intents and examples
    nlu_data['nlu'].extend(new_data_structure)

    # Write the updated NLU data back to the file
    with open(nlu_file_path, 'w', encoding='utf8') as file:
        yaml.dump(nlu_data, file, allow_unicode=True, default_flow_style=False, sort_keys=False)

# Use the function with your DataFrame and NLU file path
append_df_to_nlu(df_int, 'nlu.yml')
```

```python
import pandas as pd
import yaml
from collections import defaultdict

def append_df_to_nlu(df, nlu_file_path):
    # Create a defaultdict to hold the examples for each intent
    examples_dict = defaultdict(list)

    # Append new entries to the defaultdict
    for _, row in df.iterrows():
        intent = row['Intent']
        example = row['Examples'].strip()
        examples_dict[intent].append(example)

    # Read the existing NLU data
    with open(nlu_file_path, 'r', encoding='utf8') as file:
        nlu_data = yaml.safe_load(file) or {'nlu': []}

    # Check if 'nlu' is a key in the dictionary, as expected
    if not isinstance(nlu_data.get('nlu', []), list):
        raise ValueError("The 'nlu' key is not a list in the YAML file.")

    # Build new data structure
    new_data_structure = [{'intent': intent, 'examples': f"|\n- {('\n- '.join(examples))}"} for intent, examples in examples_dict.items()]

    # Append new intents and examples if the intent does not already exist
    for new_entry in new_data_structure:
        if not any(entry.get('intent') == new_entry['intent'] for entry in nlu_data['nlu']):
            nlu_data['nlu'].append(new_entry)

    # Write the updated NLU data back to the file
    with open(nlu_file_path, 'w', encoding='utf8') as file:
        yaml.dump(nlu_data, file, allow_unicode=True, default_flow_style=False, sort_keys=False)

# Example usage with your DataFrame
nlu_file = 'nlu.yml'  # Replace with the actual path to your NLU file
append_df_to_nlu(df_int, nlu_file)
```
