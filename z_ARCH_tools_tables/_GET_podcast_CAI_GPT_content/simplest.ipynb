```python
import pandas as pd
from pathlib import Path

# Initialize the DataFrame and save it as a .pkl file
def initialize_df():
    topics = ['Ways to search for desired jobs'] * 5
    answers = ['LinkedIn', 'Job Boards', 'Networking', 'Company Websites', 'Recruitment Agencies']
    df = pd.DataFrame({
        'a': topics,
        'b': answers
    })
    df.to_pickle("job_search_data_.pkl") # ^^^ path ^^^
    print("DataFrame initialized and saved.")

# Append new rows to the existing DataFrame and save it back as a .pkl file
def append_to_df(new_topic, new_answers):
    file_path = "job_search_data_.pkl" # ^^^ path ^^^
    
    if Path(file_path).exists():
        df = pd.read_pickle(file_path)
        new_rows = pd.DataFrame({
            'a': [new_topic] * 5,
            'b': new_answers
        })
        df = pd.concat([df, new_rows], ignore_index=True)
        df.to_pickle(file_path)  # ^^^ path ^^^
        print("New data appended and DataFrame saved.")
    else:
        print("Initialize the DataFrame first by running initialize_df().")

# Display the DataFrame from the .pkl file
def display_df():
    file_path = "job_search_data_.pkl" # ^^^ path ^^^
    df = pd.read_pickle(file_path)
    #print(df)
    return(df)

# Initialize the DataFrame
initialize_df()

# Append new data
new_topic = "Best Keywords for Job Search"
new_answers = ["Data Science", "Sound Engineering", "Signal Processing", "Python", "Audio Analysis"]
append_to_df(new_topic, new_answers)

# Display the DataFrame
display_df()
```
